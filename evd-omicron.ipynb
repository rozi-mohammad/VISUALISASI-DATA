```python
import matplotlib.pyplot as plt # visualisasi data
import seaborn as sns # visualisasi data

# output dari visualisasi data akan diarahkan ke notebook
%matplotlib inline 
```

```python
sns.heatmap(data=omicron_df.corr())
```

```python
omicron_df['location'].value_counts() # menghitung jumlah setiap Lokasi (negara)
```

```python
sns.countplot(data=omicron_df, x='location')
plt.tight_layout()
# sns.countplot?
```

```python
omicron_df['location'].value_counts().plot.pie(autopct='%1.1f%%', labels=None, legend=True)
plt.tight_layout()
```

```python
omicron_df['location'].value_counts().plot.bar()
plt.tight_layout()
plt.show()
```

```python
fig,ax = plt.subplots(nrows=2, ncols=2, figsize=(8,8))

omicron_df.perc_sequences.plot.line(ax=ax[0][0])
ax[0][0].set_title('perc_sequences')

omicron_df['num_sequences'].plot.line(ax=ax[0][1])
ax[0][1].set_title('num_sequences')

omicron_df.num_sequences_total.plot.line(ax=ax[1][0])
ax[1][0].set_title('num_sequences_total')
```

```python
omicron_df.plot()
plt.tight_layout()
```

```python
omicron_df.boxplot(by="location", figsize=(8,8))
plt.tight_layout()
```

```python
sns.scatterplot(x='perc_sequences', y='num_sequences', data=omicron_df, hue='variant')
plt.tight_layout()
```

```python
sns.violinplot(data=omicron_df, y='variant', x='perc_sequences', inner='quartile')
plt.tight_layout()
```
